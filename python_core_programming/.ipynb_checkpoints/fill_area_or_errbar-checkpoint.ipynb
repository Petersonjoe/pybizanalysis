```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2 * np.pi * x)
y2 = 0.8 * np.sin(4 * np.pi * x)
```

```python
print(x.__class__.__name__)
print(y1.__class__.__name__)
print(y2.__class__.__name__)
```

```python
# ax1, ax2, ax3是分别拿到三个子图的句柄方便操作
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(6, 6))

# 如果第三个参数不声明，则默认为0
ax1.fill_between(x, y1)
ax1.set_title('fill between y1 and 0')

ax2.fill_between(x, y1, 1)
ax2.set_title('fill between y1 and 1')

ax3.fill_between(x, y1, y2)
ax3.set_title('fill between y1 and y2')
ax3.set_xlabel('x')
fig.tight_layout()
```

```python
N = 21
x = np.linspace(0, 10, 11)
y = [3.9, 4.4, 10.8, 10.3, 11.2, 13.1, 14.1,  9.9, 13.9, 15.1, 12.5]

plt.plot(x,y)
```

```python
# 拟合返回的参数，斜率和截距
a, b = np.polyfit(x, y, deg=1)
# 拟合的函数 y=ax+b
y_est = a * x + b
y_err = x.std() * np.sqrt(1/len(x) +
                          (x - x.mean())**2 / np.sum((x - x.mean())**2))


print(len(x), x, x.std(), x.mean())
```

```python
fig, ax = plt.subplots()
ax.plot(x, y_est, '-')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.2)
ax.plot(x, y, 'o', color='tab:orange')
```

---
#### 最小二乘原理的解释

直线拟合 $y=ax+b$ 观测值$x_i (i=1...N)$被认为是精确的。所有的误差只和$y_i$相联系。

相关解释：【拟合值的标准误】https://support.minitab.com/zh-cn/minitab/18/help-and-how-to/modeling-statistics/regression/how-to/partial-least-squares/methods-and-formulas/fits-and-residuals/#residuals

拟合值标准差（拟合值SE），$y_{err}=\sqrt{{S^2}\left[\frac{1}{N}+\frac{(x_0-\bar{x})^2}{\sum^{N}_{i=1}{(x_i-\bar(x))^2}}\right]}$，其中$S^2$为均方误差。
