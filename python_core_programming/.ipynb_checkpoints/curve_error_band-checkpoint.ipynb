### Part 1

极坐标系的曲线图可以直接用`plot`来绘制
$Note: error band = 误差带$

```python
import numpy as np
from scipy.interpolate import splprep, splev

import matplotlib.pyplot as plt
from matplotlib.path import Path
from matplotlib.patches import PathPatch
```

```python
N = 400
t = np.linspace(0,2*np.pi, N)
r = 0.5 + np.cos(t)
x,y = r*np.cos(t), r*np.sin(t)

fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

在上面的计算中，相应的函数具体公式是什么？
首先，$t$是一个参数，且$t$是[0, $2\pi$]内的序列。`np.linspace`应该可以按均匀间隔产生一定数目，一定范围内的数。

```python
a = np.linspace(0, 9, 10)
print(a)
```

所以具体公式可以是：
$z=\Bigg\{\begin{array}{lr} r=0.5+\cos(t) & \\ x=r\cos(t) & \\ y=r\sin(t) \end{array}$

也就是说，
$\begin{array}{lr} x=0.5\cos(t) + \cos^2(t) & \\ y=0.5\sin(t) + \cos(t)\sin(t) \end{array}$

同时，很简单可以知道，
$\begin{array}{lr} x^2+y^2=r^2=0.25+0.5\cos(t)+cos^2(t) & \\ \Rightarrow x^2+y^2=r^2=0.25+0.5\cos(t)+(x-0.5\cos(t)) & \\ \Rightarrow x^2-x+y^2=0.25 \end{array}$

```python
err = 0.05 * np.sin(2 * t) ** 2 + 0.04 + 0.02 * np.cos(9 * t + 2)

tck, u = splprep([x, y], s=0)
dx, dy = splev(u, tck, der=1)
l = np.hypot(dx, dy)
nx = dy / l
ny = -dx / l

xp = x + nx * err
yp = y + ny * err
xn = x - nx * err
yn = y - ny * err

vertices = np.block([[xp, xn[::-1]],
                     [yp, yn[::-1]]]).T
codes = Path.LINETO * np.ones(len(vertices), dtype=Path.code_type)
codes[0] = codes[len(xp)] = Path.MOVETO
path = Path(vertices, codes)

patch = PathPatch(path, facecolor='C0', edgecolor='none', alpha=0.3)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.add_patch(patch)
plt.show()
```
